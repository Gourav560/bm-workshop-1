```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
fig = plt.figure(figsize = (8,8), dpi=100)
m = Basemap(projection = 'lcc',resolution ='h', width = 7E6, height = 7E6, lat_0=21.43,lon_0=87.01)
m.etopo(scale=0.5,alpha=0.5)
m.bluemarble(scale=0.5)
#28.779446513403524, 77.05195739785829
x,y = m(77.05195739785829, 28.779446513403524)
plt.plot(x,y, 'ok', markersize=5)
plt.text(x,y, 'My address', fontsize=12);
plt.savefig('myaddress.png', dpi=600, bbox_inches='tight')
plt.show()
```
